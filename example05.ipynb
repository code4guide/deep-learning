```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
import math

fan_in, fan_out = 100, 50

# 1) 그냥 정규분포 N(0, 1)
w_normal = torch.empty(fan_out, fan_in)
nn.init.normal_(w_normal, mean=0.0, std=1.0)

# 2) Xavier Uniform
w_xavier = torch.empty(fan_out, fan_in)
bound = math.sqrt(6.0 / (fan_in + fan_out))
nn.init.uniform_(w_xavier, -bound, bound)

w_normal = w_normal.flatten().numpy()
w_xavier = w_xavier.flatten().numpy()

plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.hist(w_normal, bins=50)
plt.title("Normal(0, 1)")

plt.subplot(1, 2, 2)
plt.hist(w_xavier, bins=50)
plt.title("Xavier Uniform")

plt.show()
```
